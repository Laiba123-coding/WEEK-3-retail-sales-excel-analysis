Convert Date Columns

```python
df["PO Sent to Vendor Date"] = pd.to_datetime(df["PO Sent to Vendor Date"], format='%d-%b-%y', errors='coerce')
df["Scheduled Delivery Date"] = pd.to_datetime(df["Scheduled Delivery Date"], format='%d-%b-%y', errors='coerce')
df["Delivered to Client Date"] = pd.to_datetime(df["Delivered to Client Date"], format='%d-%b-%y', errors='coerce')
```

```python
df["PO Sent to Vendor Date"] = pd.to_datetime(
    df["PO Sent to Vendor Date"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df["PO Sent to Vendor Date"].dtype
```

```python
df["PO Sent to Vendor Date"].head(20)
```

Convert Numeric Columns

Sometimes numbers are stored as text.

Check

```python
df.dtypes
```

```python
date_columns = ['PQ First Sent to Client Date', 'Scheduled Delivery Date', 'Delivered to Client Date', 'Delivery Recorded Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
df[date_columns].dtypes
```

```python
df["Freight Cost (USD)"] = pd.to_numeric(df["Freight Cost (USD)"], errors="coerce")

df["Weight (Kilograms)"] = pd.to_numeric(df["Weight (Kilograms)"], errors="coerce")

df["Unit Price"] = pd.to_numeric(df["Unit Price"], errors="coerce")

df["Pack Price"] = pd.to_numeric(df["Pack Price"], errors="coerce")

df["Line Item Value"] = pd.to_numeric(df["Line Item Value"], errors="coerce")
```

```python
df.isnull().sum()
```

Fill Missing Values

Numeric

```python
df["Shipment Mode"] = df["Shipment Mode"].fillna(df["Shipment Mode"].mode()[0])
```

```python
df["Line Item Insurance (USD)"] = df["Line Item Insurance (USD)"].fillna(
    df["Line Item Insurance (USD)"].mean()
)
```

Text columns

```python
df["Dosage"] = df["Dosage"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
print(df["PO Sent to Vendor Date"].dtype)
print(df["PO Sent to Vendor Date"].isna().sum())
```

```python
df[df["PO Sent to Vendor Date"].isna()].head()
```

```python
df = df.drop(columns=["PO Sent to Vendor Date"])
```

```python
# Convert to numeric
df["Weight (Kilograms)"] = pd.to_numeric(
    df["Weight (Kilograms)"], errors="coerce"
)

df["Freight Cost (USD)"] = pd.to_numeric(
    df["Freight Cost (USD)"], errors="coerce"
)
```

```python
df["Weight (Kilograms)"] = df["Weight (Kilograms)"].fillna(
    df["Weight (Kilograms)"].mean()
)

df["Freight Cost (USD)"] = df["Freight Cost (USD)"].fillna(
    df["Freight Cost (USD)"].mean()
)
```

```python
df.info()
```

```python
median_date = df["PQ First Sent to Client Date"].median()

df["PQ First Sent to Client Date"] = df["PQ First Sent to Client Date"].fillna(median_date)
```

Handling Missing Values:

*   Filled missing values in Shipment Mode using the mode (most frequent value).
*   Filled missing values in Dosage with "Unknown".
*   Filled missing values in Line Item Insurance (USD) using the column mean.
*    Removed the PO Sent to Vendor Date column because it contained a large number of missing values and was not required for the delivery performance or cost analysis.
